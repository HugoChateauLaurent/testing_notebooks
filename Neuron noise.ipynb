```python
import nengo

model = nengo.Network()
with model:
    a = nengo.Ensemble(n_neurons=10, dimensions=1)
    
    dist = nengo.utils.distributions.Uniform(-1, 1)
    noise = nengo.Node(lambda t: dist.sample(10))
    nengo.Connection(noise, a.neurons)
    
    probe = nengo.Probe(a, synapse=0.03)
    
sim = nengo.Simulator(model)
sim.run(1)
plot(sim.trange(), sim.data[probe])
```
